# Pivot

```python
#create my own dataset

sample_id = np.arange(1,6).repeat(2)
target = [i for i in "AB"*5]
ph = np.random.normal(7, 0.5, 10).round(1) #normal distribution with mu = 7 and sigma = 0.5, rounded to single decimal
viral_count = np.random.randint(500, 10000, 10)
waste_site = ["".join(choice(string.ascii_letters) for i in range(10)) for x in range(10)]
flow_rate = np.random.randint(50,150, 10)

data_dict = {"sample_id": sample_id,
             "target":target,
             "ph":ph,
             "viral_count":viral_count,
             "waste_site": waste_site,
             "flow_rate": flow_rate
            }

df = pd.DataFrame(data_dict)
```

```python
#view dataframe
df
```

```python
#explore pivot function
df.pivot?
```

```python
#apply pivot function while omitting index
df.pivot(columns = "target")
```

```python
#set index to sample_id for the pivot
df.pivot(columns = "target", index = "sample_id")
```

```python
#select only desired columns 
df.pivot(columns = "target", index = "sample_id", values = ["ph", "viral_count", "flow_rate"])
```

```python
#some magic to rename and collapse multi-index columns
df_piv = df.pivot(columns = "target", index = "sample_id", values = ["ph", "viral_count", "flow_rate"])
df_piv.columns = ["_".join(col) for col in df_piv.columns.values]
df_piv
```

# Wide to long transform: melt

```python
df_long = pd.DataFrame({
        "student":
            ["Andy", "Bernie", "Cindy", "Deb",
             "Andy", "Bernie", "Cindy", "Deb",
             "Andy", "Bernie", "Cindy", "Deb"],
        "school":
            ["Z", "Y", "Z", "Y",
             "Z", "Y", "Z", "Y",
             "Z", "Y", "Z", "Y"],
        "class":
            ["english", "english", "english", "english",
             "math", "math", "math", "math",
             "physics", "physics", "physics", "physics"],
        "grade":
            [10, 100, 1000, 10000,
             20, 200, 2000, 20000,
             30, 300, 3000, 30000]
        
})
```

```python
df_long
```

```python
df_wide = df_long.pivot_table(index = ["school", "student"],
                             columns = "class", 
                             values = "grade").reset_index()
                             
df_wide
```

```python
df_wide.melt?
```

```python
df_wide.melt(id_vars = ["school", "student"],
             value_vars = ["english","math","physics"],
             var_name = "class",
            value_name = "grade")
```
